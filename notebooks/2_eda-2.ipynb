### Objective
This notebook performs exploratory data analysis on Kenya's youth
unemployment dataset (1991–2025). We examine long-term trends,
decade patterns, year-on-year changes, and rolling averages to
extract meaningful insights from the data.

```python
# ============================================================
# SECTION 1 — Import Libraries & Load Data
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go

# Load clean dataset
df = pd.read_csv('kenya_unemployment_clean.csv')

print("Dataset loaded successfully")
print(f"Shape: {df.shape}")
print(f"\nColumns: {df.columns.tolist()}")
print(f"\nFirst 5 rows:")
print(df.head())
```

---
## Section 2 — Long-Term Trend Analysis

We plot the unemployment rate over time alongside the 3-year rolling
average to distinguish the structural trend from short-term fluctuations.

```python
# ============================================================
# SECTION 2 — Long-Term Trend Analysis
# ============================================================

fig = go.Figure()

# Actual unemployment rate
fig.add_trace(go.Scatter(
    x=df['Year'],
    y=df['Unemployment_Rate'],
    mode='lines+markers',
    name='Unemployment Rate',
    line=dict(color='royalblue', width=2),
    marker=dict(size=6),
    hovertemplate='Year: %{x}<br>Rate: %{y:.3f}%<extra></extra>'
))

# Rolling average
fig.add_trace(go.Scatter(
    x=df['Year'],
    y=df['Rolling_Avg'],
    mode='lines',
    name='3-Year Rolling Average',
    line=dict(color='red', width=2, dash='dash'),
    hovertemplate='Year: %{x}<br>Rolling Avg: %{y:.3f}%<extra></extra>'
))

fig.update_layout(
    title=dict(
        text='Kenya Youth Unemployment Rate with Rolling Average (1991–2025)',
        font=dict(size=16),
        x=0.5,
        xanchor='center'
    ),
    xaxis=dict(
        title='Year',
        tickmode='array',
        tickvals=[1991, 1995, 2000, 2005, 2010, 2015, 2020, 2025],
        showgrid=True,
        gridcolor='lightgrey'
    ),
    yaxis=dict(
        title='Unemployment Rate (%)',
        showgrid=True,
        gridcolor='lightgrey'
    ),
    hovermode='x unified',
    plot_bgcolor='white',
    paper_bgcolor='white',
    legend=dict(x=0.01, y=0.99)
)

fig.show()
print("Chart rendered")
```

### Insight — Long-Term Trend

The chart reveals a clear structural upward trend in Kenya's youth
unemployment from 1991 to 2025. The 3-year rolling average confirms
this is not a short-term fluctuation but a persistent long-term pattern.

Observed Pattern:
- 1991–2000: Relatively stable unemployment between 6.2% and 6.8%
- 2000–2010: Gradual increase begins, crossing 7% for the first time
- 2010–2019: Accelerated growth, unemployment approaches double digits
- 2020–2022: Sharp spike driven by COVID-19 economic disruption
- 2022–2025: Slight decline but remains historically high above 15%

Economic Interpretation:
The sustained upward trend reflects a phenomenon economists call
"jobless growth" — where Kenya's GDP has expanded over the decades
without generating proportional employment opportunities for young people.

Three structural forces drive this:

1. Youth Population Pressure
   Kenya has one of the youngest populations in the world. The youth
   cohort (15–24) has grown faster than the economy's capacity to
   absorb new workers. Each year, hundreds of thousands of young
   Kenyans enter a labour market that lacks sufficient formal jobs.

2. Skills Mismatch
   Kenya's education system has historically emphasized academic
   credentials over practical, market-relevant skills. Graduates
   enter the workforce without the technical competencies demanded
   by employers — particularly in technology, engineering, and
   data-driven industries. This creates a paradox where jobs exist
   but qualified young applicants do not.

3. Structural Economic Composition
   Kenya's economy remains heavily dependent on agriculture and the
   informal sector, neither of which provides stable, formal
   employment for educated youth. The formal private sector — which
   could absorb skilled graduates — remains underdeveloped relative
   to the scale of youth unemployment.

Implication:
The trend will not reverse on its own. Without deliberate structural
reforms in education, industrial policy, and youth entrepreneurship
support, Kenya's youth unemployment crisis will deepen — as confirmed
by our ARIMA forecast projecting rates above 16% by 2030.


```python
# ============================================================
# SECTION 3 — Year-on-Year Change Analysis
# ============================================================

# Drop NaN for Rate_Change
df_change = df.dropna(subset=['Rate_Change'])

fig = go.Figure()

# Color bars based on positive/negative change
colors = ['crimson' if x > 0 else 'steelblue' for x in df_change['Rate_Change']]

fig.add_trace(go.Bar(
    x=df_change['Year'],
    y=df_change['Rate_Change'],
    marker_color=colors,
    name='Year-on-Year Change',
    hovertemplate='Year: %{x}<br>Change: %{y:.3f}%<extra></extra>'
))

fig.add_hline(y=0, line_color='black', line_width=1)

fig.update_layout(
    title=dict(
        text='Year-on-Year Change in Kenya Youth Unemployment (1992–2025)',
        font=dict(size=16),
        x=0.5,
        xanchor='center'
    ),
    xaxis=dict(
        title='Year',
        tickmode='array',
        tickvals=[1992, 1995, 2000, 2005, 2010, 2015, 2020, 2025],
        showgrid=True,
        gridcolor='lightgrey'
    ),
    yaxis=dict(
        title='Change in Unemployment Rate (%)',
        showgrid=True,
        gridcolor='lightgrey'
    ),
    hovermode='x unified',
    plot_bgcolor='white',
    paper_bgcolor='white'
)

fig.show()
print("Chart rendered")
```

### Insight — Year-on-Year Change

The bar chart reveals the volatility pattern in Kenya's youth
unemployment, showing which years saw increases (red) and
decreases (blue).

Observed Pattern:
- 1990s: Small fluctuations in both directions — relatively stable period
- 2001–2003: Notable increase following global economic slowdown
- 2008–2009: Sharp spike coinciding with the Global Financial Crisis
- 2011–2019: Mixed signals — alternating increases and decreases
- 2020–2021: Largest single-year increase on record driven by COVID-19
- 2023–2025: Consecutive decreases suggesting post-pandemic recovery

Economic Interpretation:
The volatility pattern reveals that Kenya's youth unemployment is
highly sensitive to external economic shocks — a characteristic of
economies with weak structural buffers.

Three key economic events stand out:

1. Global Financial Crisis (2008–2009)
   The GFC triggered a collapse in global demand, foreign investment,
   and remittances — all critical to Kenya's economy. Export-dependent
   sectors contracted sharply, eliminating entry-level jobs that young
   workers disproportionately occupy. Kenya's lack of a formal social
   safety net meant youth bore the heaviest burden of the downturn.

2. COVID-19 Pandemic (2020–2022)
   The pandemic produced the largest single-year unemployment spike
   in the dataset. Kenya's tourism sector — a major employer of young
   workers — collapsed entirely. Lockdowns devastated the informal
   economy where most youth employment is concentrated. The spike
   reflects both direct job losses and the failure of formal systems
   to provide alternative employment pathways during the crisis.

3. Post-Pandemic Recovery (2023–2025)
   The consecutive decreases in 2023–2025 suggest partial recovery.
   However rates remain historically elevated above 15%, indicating
   that recovery has been insufficient to restore pre-pandemic levels
   let alone address the underlying structural trend.

Implication:
Kenya's youth employment is structurally fragile — highly exposed
to global shocks with limited domestic buffers. Building resilience
requires diversifying the economy beyond tourism and agriculture
into technology, manufacturing, and knowledge industries that can
sustain employment even during global downturns.

```python
# ============================================================
# SECTION 4 — Decade Comparison Analysis
# ============================================================

decade_stats = df.groupby('Decade')['Unemployment_Rate'].agg([
    ('Average', 'mean'),
    ('Minimum', 'min'),
    ('Maximum', 'max'),
    ('Std_Dev', 'std')
]).round(3).reset_index()

print("Unemployment statistics by decade:")
print(decade_stats)

# Plot
fig = go.Figure()

fig.add_trace(go.Bar(
    x=decade_stats['Decade'],
    y=decade_stats['Average'],
    name='Average Rate',
    marker_color='steelblue',
    text=decade_stats['Average'],
    textposition='outside',
    texttemplate='%{text:.2f}%',
    hovertemplate='Decade: %{x}<br>Average: %{y:.3f}%<extra></extra>'
))

fig.add_trace(go.Scatter(
    x=decade_stats['Decade'],
    y=decade_stats['Maximum'],
    mode='markers',
    name='Peak Rate',
    marker=dict(color='crimson', size=12, symbol='diamond'),
    hovertemplate='Decade: %{x}<br>Peak: %{y:.3f}%<extra></extra>'
))

fig.update_layout(
    title=dict(
        text='Kenya Youth Unemployment by Decade — Average and Peak Rates',
        font=dict(size=16),
        x=0.5,
        xanchor='center'
    ),
    xaxis=dict(title='Decade', showgrid=False),
    yaxis=dict(
        title='Unemployment Rate (%)',
        showgrid=True,
        gridcolor='lightgrey',
        range=[0, decade_stats['Maximum'].max() + 3]
    ),
    hovermode='x unified',
    plot_bgcolor='white',
    paper_bgcolor='white',
    legend=dict(x=0.01, y=0.99)
)

fig.show()
print("Chart rendered")
```

### Insight — Decade Comparison

Decade-by-decade analysis reveals a consistent and accelerating
upward pattern in Kenya's youth unemployment.

Observed Pattern:
- 1990s: Lowest average at 6.55% — most stable decade
- 2000s: Moderate increase to average 7.13% — gradual deterioration
- 2010s: Significant jump to average 9.87% — structural shift occurs
- 2020s: Highest average at 14.92% — pandemic permanently elevated rates

Economic Interpretation:
Each decade tells a distinct economic story:

1. The 1990s — Structural Adjustment Era
   Kenya underwent IMF and World Bank structural adjustment programs
   during this period. While these reforms stabilized macroeconomic
   indicators, they also reduced public sector employment —
   historically a major absorber of educated youth. The relatively
   low unemployment masked a growing structural vulnerability as
   public hiring slowed and the private sector was not yet developed
   enough to compensate.

2. The 2000s — Devolution and Growing Inequality
   Kenya's economy grew steadily in the 2000s driven by mobile
   technology, financial services, and construction. However this
   growth was concentrated in Nairobi and a few urban centers,
   leaving rural youth — the majority of Kenya's young population —
   largely excluded. GDP growth without geographic equity produced
   rising unemployment despite positive economic headlines.

3. The 2010s — Population Surge Outpaces Growth
   Kenya's youth population grew faster than at any previous point
   during this decade. The labour market simply could not absorb
   the volume of new entrants. Additionally Kenya's education system
   expanded rapidly — producing more graduates — without a
   corresponding expansion in formal employment opportunities.
   This decade marks the point where the skills mismatch became
   a full-scale structural crisis.

4. The 2020s — Pandemic Amplifies Structural Weakness
   COVID-19 did not create Kenya's youth unemployment crisis — it
   exposed and amplified it. The pandemic eliminated the informal
   sector jobs that had been masking the true scale of structural
   unemployment. The 2020s average of 14.92% represents not just
   pandemic disruption but the convergence of decades of accumulated
   structural failures in education, industrial policy, and
   labour market development.

Implication:
The widening gap between decade averages is not accidental. It
reflects compounding structural failures that worsen each decade
without intervention. The 2020s peak of 15.49% is more than double
the 1990s average — a trajectory that our ARIMA model projects
will continue unless deliberate policy action is taken.
